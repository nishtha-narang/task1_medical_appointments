```python
# Import the libraries
import os
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings ('ignore')
```

```python
# Load the dataset
df = pd.read_csv ('Medical_Appointments.csv')
df.head()
```

```python
#Check for rows*columns in the data
df.shape
```

```python
#Segregate the true numeric and true object columns
df.info()
```

```python
#Check for values in each column to get an overview
for i in df:
    print (df[i].value_counts())
```

```python
#Look for duplicate values in the dataset
df.duplicated().sum()
```

```python
#'-1' in age being an unrealistic value is replaced with null
df['Age'] = df['Age'].replace(-1, np.nan)
```

```python
#Look for missing values in the dataset
df.isnull().sum()
```

```python
#Impute null value in age column with median
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
#Convert dtype to 'int64' for easy display
df['PatientId'] = df['PatientId'].astype('int64')
```

```python
#Look for top 5 values 
df.head()
```

```python
#Rename columns for standardisation
df.rename(columns={
    'PatientId': 'patient_id',
    'AppointmentID': 'appointment_id',
    'ScheduledDay': 'scheduled_day',
    'AppointmentDay': 'appointment_day',
    'Hipertension': 'hypertension',  
    'Handcap': 'handicap',           
    'No-show': 'no_show'
}, inplace=True)
```

```python
#Convert column headers to lowercase for standardisation
df.columns = df.columns.str.lower()
```

```python
#Convert dtype of date columns 
df['scheduled_day'] = pd.to_datetime(df['scheduled_day'])
df['appointment_day'] = pd.to_datetime(df['appointment_day'])
```

```python
#Changed dtypes
df.info()
```

```python
#Map gender values as required
df['gender'] = df['gender'].map({'M': 1, 'F': 0}).astype(int)
```

```python
#Convert to int dtype
df['no_show'] = df['no_show'].map({'No': 0, 'Yes': 1}).astype(int)
```

```python
#Encode neighbourhood column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['neighbourhood'] = le.fit_transform(df['neighbourhood'])
```
